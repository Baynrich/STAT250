# Task 1

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(27)

uniform_samples = [np.random.uniform() for i in range(500)]
def twoexp_inv(x, lmbda, eta):
    return eta - ((np.log(1-x)) / lmbda)

# 1.1
distribution_samples = [twoexp_inv(x, 1, 1) for x in uniform_samples]
print(min(distribution_samples))
bins = [1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6, 6.5, 7]
plt.hist(distribution_samples, bins=bins)
plt.show()
```

```python
# 1.2
uniform_samples_2a = [np.random.uniform() for i in range(100)]
distribution_samples_2a = [twoexp_inv(x, 1, 10) for x in uniform_samples_2a]
plt.hist(distribution_samples_2a)
plt.show()

uniform_samples_2b = [np.random.uniform() for i in range(1000)]
distribution_samples_2b = [twoexp_inv(x, 1, 10) for x in uniform_samples_2b]
plt.hist(distribution_samples_2b)
plt.show()

uniform_samples_2c = [np.random.uniform() for i in range(1000)]
distribution_samples_2c = [twoexp_inv(x, 10, 1) for x in uniform_samples_2c]
plt.hist(distribution_samples_2c)
plt.show()
```

# Task 2

```python
def resolve_uniform(x):
    if x < 0.0625:
        return 0
    if x < 0.3125:
        return 1
    if x < 0.6875:
        return 2
    if x < 0.9375:
        return 3
    return 4
```

```python
draw_uniform = [np.random.uniform() for i in range(500)]
resolveds = [resolve_uniform(x) for x in draw_uniform]
plt.hist(resolveds)
plt.show()
```

# Task 3

```python
def f_over_cg(y, a, b):
    return ((y/(a-1))**(a-1)) * ((1-y)/(b-1))**(b-1) * ((a+b-2)**(a+b-2))

def generate_accept_reject_samples(n, a, b):
    samples = []
    while len(samples) < n:
        y = np.random.uniform()
        prob = f_over_cg(y, a, b)
        u = np.random.uniform()
        if u < prob:
            samples.append(y)
    return samples

samples_a = generate_accept_reject_samples(100, 2, 3)
plt.hist(samples_a)
plt.show()

samples_b = generate_accept_reject_samples(1000, 2, 3)
plt.hist(samples_b)
plt.show()

samples_c = generate_accept_reject_samples(1000, 5, 2)
plt.hist(samples_c)
plt.show()

        
        
```
